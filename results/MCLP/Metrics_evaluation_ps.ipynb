```python
import setup_path
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import string
from src.utils.graphs import error_line_graph
```

```python
labels = ['UP', 'UP_LR', 'UP_WS', 'UP_WS_LR']
labels1 = ['UP', 'UP_LR', 'UP_WS', 'UP_LR_WS']
markers = ['.', '.', '*', '*']
lss = ['-', '--', '-', '--']

metrics = ['r', 'probs', 'qaoa_FR', 'n_fev']
data_list = []

#file_name = "data_ps_[3]_[3].txt"
file_name = "data_ps_3_3_1000.txt"
with open(file_name, "r") as file:
    loaded_data = eval(file.read())

for mtr in metrics:
    data_list.append([np.array(loaded_data[label][mtr]).T for label in labels]) 
```

```python
print(loaded_data['UP'].keys())
```

```python
ps = [1, 20, 40, 60, 80, 100]
x = [ps] * len(labels)
raw_data = data_list
labels = labels
title = None
y_labels = [r'$r$', r'$P(x*)$', 'FR', 'Function Evaluations']
x_label = 'p'
y_lim = None
font_s=28
label_s=21

fig, axs = plt.subplots(2, 2, figsize=(15, 12))
axs = axs.flatten()
for i in range(4):
    error_line_graph(axs[i], x, raw_data[i], labels1, title, x_label, y_labels[i], markers, lss)
    if i%2!=0:
        axs[i].yaxis.tick_right()  
        axs[i].yaxis.set_label_position("right") 
    axs[i].minorticks_off()
    if i <2:
        axs[i].tick_params(bottom=True, labelbottom=False)
        
    else:
        axs[i].set_xticks(ps)
        axs[i].set_xticklabels(ps)
        axs[i].set_xlabel(x_label, fontsize=font_s)
    axs[i].tick_params(labelsize=label_s)
    axs[i].set_ylabel(y_labels[i], fontsize=font_s)
    axs[i].grid(linewidth=0.2)

axs[0].set_ylim([0.66, 1.04])
axs[1].set_ylim([-0.05, 1.09])
axs[2].set_ylim([0.05, 1.09])
axs[3].set_ylim([-10, 1100])

letters = [f"({l})" for l in string.ascii_lowercase[:len(axs)]]
for ax, letter in zip(axs, letters):
    ax.text(0.06, 0.98, letter, transform=ax.transAxes,
            fontsize=21, fontweight='bold', va='top', ha='left')

fig.legend(
    labels1,
    frameon=True,
    fontsize=font_s,
    loc='upper center',
    bbox_to_anchor=(0.5, 1),
    ncol=4
)

plt.tight_layout(rect=[0, 0, 1, 0.94])  # space for legend
plt.savefig('Figure_3.pdf', dpi=300)
plt.show()
```
