```python
import numpy as np
from collections import defaultdict
import json
import random
```

```python
seed = 341448
if seed is not None:
    np.random.seed(seed)
    random.seed(seed)
```

```python
def generate_grid(rows, cols, p, seed=None):
    """
    Generate a grid with n unique non-zero integers placed randomly.
    Remaining cells are filled with 0.
    """
    total = rows * cols
    n = int(p * total)
        
    if n > total:
        raise ValueError("n cannot be greater than total grid cells.")
    if n < 0:
        raise ValueError("n must be non-negative.")

    eleg_fac_loc = [1 for _ in range(n)] #eligible sites

    # Fill the rest with zeros
    eleg_fac_loc += [0] * (total - n)

    # Shuffle to randomize positions
    random.shuffle(eleg_fac_loc)

    population = [np.random.randint(1, 6) for _ in range(total)]

    return eleg_fac_loc, population, n
```

```python
seeds = [3246, 2096, 3103, 435, 364, 6834, 4399, 2034, 9123, 2385, 3541, 5264, 2137, 3542, 23504, 35023, 7304, 30461, 37894, 334542354]
```

```python
mclp_instances = defaultdict(list)
n_m_vals = [[2, 2], [2, 3], [2, 4], [3, 3], [3, 4], [4, 4]]
p = 0.5
num_inst = 20

for n, m in n_m_vals:
    for k in range(num_inst):
        grid = generate_grid(n, m, p)
        mclp_instances[str([n, m])].append(grid)

print(mclp_instances)
```

```python
try:
    with open("data_MCLP.txt", "w", encoding="utf-8") as file:
        json.dump(mclp_instances, file, indent=4)
    print("Data successfully written to data.txt")
except IOError as e:
    print(f"Error writing file: {e}")
```
