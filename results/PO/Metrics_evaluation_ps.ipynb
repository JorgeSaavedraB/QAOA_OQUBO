```python
import setup_path
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import string
from src.utils.graphs import error_line_graph
```

```python
labels = ['SV_WS_LR', 'UP_WS_LR', 'SV_WS', 'UP_WS', 'SV_LR', 'UP_LR', 'SV', 'UP']
markers = ['*', '*', '*', '*', '.', '.', '.', '.']
lss = ['--', '-', '--', '-', '--', '-', '--', '-']
metrics = ['r', 'probs', 'qaoa_FR', 't_total']
data_list = []

file_name = "data_N16_r0-5_ps.txt"
with open(file_name, "r") as file:
    loaded_data = eval(file.read())

for mtr in metrics:
    data_list.append([np.array(loaded_data[label][mtr]).T for label in labels]) 
```

```python
ps = [1, 2, 4, 8, 16, 32]
x = [ps] * len(labels)
raw_data = data_list
labels = labels
title = None
y_labels = ['r', 'P(x*)', 'FR', 'Tiempo total de ejecución (s)']
x_label = 'p'
y_lim = None

fig, axs = plt.subplots(2, 2, figsize=(15, 15))
axs = axs.flatten()
for i in range(4):
    error_line_graph(axs[i], x, raw_data[i], labels, title, x_label, y_labels[i], markers, lss)
    if i%2!=0:
        axs[i].yaxis.tick_right()  
        axs[i].yaxis.set_label_position("right") 
    axs[i].set_xscale('log')
    axs[i].minorticks_off()
    if i <2:
        axs[i].tick_params(labelbottom=False)
        axs[i].tick_params(bottom=False, labelbottom=False)
        
    else:
        axs[i].set_xticks(ps)
        axs[i].set_xticklabels(ps)
        axs[i].set_ylim(y_lim)
        axs[i].set_xlabel(x_label, fontsize=24)
    axs[i].set_ylabel(y_labels[i], fontsize=24)


letters = [f"({l})" for l in string.ascii_lowercase[:len(axs)]]
for ax, letter in zip(axs, letters):
    ax.text(0.02, 0.99, letter, transform=ax.transAxes,
            fontsize=20, fontweight='bold', va='top', ha='left')

fig.legend(
    labels,
    frameon=True,
    fontsize=24,
    loc='upper center',
    bbox_to_anchor=(0.5, 1),
    ncol=4
)

plt.tight_layout(rect=[0, 0, 1, 0.93])  # space for legend
plt.savefig('ps_16.pdf')
plt.show()
```
